```python
fig=plt.figure(figsize=(10,6))
```

```python
colors=['red','blue','green','orange','black']
```

```python
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    label = str(1948+i)
    plt.plot(subset['DATE'],subset['VALUES'],c=colors[i],label=label)    
    plt.xticks(rotation=45)
plt.legend(loc='best')#自动放到合适的地方
```
